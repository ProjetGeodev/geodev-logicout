```python
#We prepare the database to allow those to be count correctly and take only the entries that have valid candidates
ranked_simulations['mutualisations']=ranked_simulations['mutualisations'].replace(np.nan,'[]',regex=True)
ranked_simulations['mutualisations'] = ranked_simulations['mutualisations'].map(lambda x: ast.literal_eval(x))
ranked_simulations['nbr'] = ranked_simulations['mutualisations'].map(lambda x: len(x))
ranked_simulations['jaccard_null'] = ranked_simulations['mutualisations'].map(lambda x: [y for y in x if y[1]==0])
ranked_simulations['nbr_jaccard_null'] = ranked_simulations['jaccard_null'].map(lambda x: len(x))

ranked_simulations.describe()
```
