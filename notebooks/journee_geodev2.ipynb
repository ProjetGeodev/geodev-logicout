```python
import sys  
sys.path.insert(0, '../scripts')
import numpy as np
import use_data
import pandas as pd
import geopandas as gpd
import aux_plot
import warnings
import ast
import random
warnings.filterwarnings('ignore')
ranked_simulations = pd.read_csv("../data/raw/ranked_mutualisations.csv")
geo_df= use_data.create_gdf('simulations_reel_gdf.csv','cheflieu')
```

```python
#pick by chance
itineraries= ranked_simulations['id_simulation'].values.tolist()
random_itinerary= random.choice(itineraries)
random_itinerary
```

```python
#pick the best placed mutualisation
itinerary2=ranked_simulations[ranked_simulations['id_simulation']==random_itinerary]
itinerary2['mutualisations']=itinerary2['mutualisations'].map(lambda x: ast.literal_eval(x))
itinerary2=int(itinerary2.values[0][1][0][0])
print(random_itinerary,itinerary2)
```

```python

it1_ = use_data.get_itineraire(random_itinerary,geo_df)
it2_ = use_data.get_itineraire(itinerary2,geo_df)
```
