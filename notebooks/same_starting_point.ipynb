```python
import sys  
sys.path.insert(0, '../scripts')
import numpy as np
import indexes, use_data, IsInclude
import pandas as pd
from shapely import wkt,MultiLineString
import ast
import warnings
warnings.filterwarnings('ignore')
import aux_plot
```

```python
parameters = {
'dist':100000,
'buffer_hull':1000,
"type":1,
"area" : np.pi*100000**2
}


geo_df = use_data.create_gdf('simulations_reel_gdf.csv','start')
ranked_simulations = pd.read_csv("../data/raw/ranked_mutualisations.csv")
```

```python
#We prepare the database to allow those to be count correctly and take only the entries that have valid candidates
ranked_simulations['mutualisations']=ranked_simulations['mutualisations'].replace(np.nan,'[]',regex=True)
ranked_simulations['mutualisations'] = ranked_simulations['mutualisations'].map(lambda x: ast.literal_eval(x))
ranked_simulations['n_candidates'] = ranked_simulations['mutualisations'].map(lambda x: len(x))
ranked_simulations=ranked_simulations[ranked_simulations['n_candidates'] != 0]
```

```python
ranked_simulations['n_candidates'].describe()
```

Now we will inspect how many of those candidates have an index of mutualisation = 0

```python
ranked_simulations['candidates_0']=ranked_simulations['mutualisations'].map(lambda x: [y for y in x if y[5]==0])
ranked_simulations['n_candidates_0'] = ranked_simulations['candidates_0'].map(lambda x: len(x))
ranked_simulations_index_0 = ranked_simulations[ranked_simulations['n_candidates_0'] != 0]
```

```python
ranked_simulations_index_0[['n_candidates','n_candidates_0']].describe()
```
